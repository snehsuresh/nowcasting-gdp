```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd
# Prepare data
data = pd.read_csv('data/feature_engineered_data.csv')
data.reset_index(drop=True, inplace=True)
```

```python
data.head()
```

```python
# Define the predictors and target
predictors = ["Lagged_Sentiment_Index", "GDP_Growth_Rate"]
target = "GDP_Growth_Rate"
```

```python
predictions = []
actuals = []
```

```python
# Pseudo out-of-sample forecasting
for i in range(4, len(data)):
    # Train on data up to the current point
    train_data = data.iloc[:i]
    test_data = data.iloc[i : i + 1]

    # Define training and testing sets
    X_train = train_data[predictors]
    y_train = train_data[target]
    X_test = test_data[predictors]
    y_test = test_data[target]

    # Print debug info for the current step
    print(f"Iteration {i}")
    print("Train Data:")
    print(train_data.tail())
    print("\nTest Data:")
    print(test_data)

    # Train the model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Store the results
    predictions.append(y_pred[0])
    actuals.append(y_test.values[0])
```

```python
# Evaluate the model
mse = mean_squared_error(actuals, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(actuals, predictions)

# Display results
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
```
